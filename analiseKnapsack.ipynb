```python
import pandas as pd
import matplotlib.pyplot as plt

quant_geracoes = 500
quant_segmentos = 100 # (iteracoes do algoritmo - mesmos items, populao diferente)
df = pd.read_csv('medias.csv')
segmentos = [df.iloc[i:i+quant_geracoes] for i in range(0, len(df), quant_geracoes)]

colors = ['blue']

plt.figure(figsize=(8, 4))

for i in range(quant_segmentos):
    novo_indice = range(quant_geracoes)  
    plt.plot(novo_indice, segmentos[i]['media_peso'], color=colors[i % len(colors)], label=f'Populao {i+1}')
# plt.ylim(bottom=60)
plt.title('Mdia do Peso por Populao')
plt.xlabel('Gerao')
plt.ylabel('Peso Mdio')
# plt.xticks(range(quant_geracoes), range(1, quant_geracoes + 1))
plt.show()

plt.figure(figsize=(8, 4))

for i in range(quant_segmentos):
    novo_indice = range(quant_geracoes)  
    plt.plot(novo_indice, segmentos[i]['media_valor'], color=colors[i % len(colors)], label=f'Populao {i+1}')

plt.title('Mdia do Valor por Populao')
plt.xlabel('Gerao')
plt.ylabel('Valor Mdio')
# plt.xticks(range(quant_geracoes), range(1, quant_geracoes + 1))
plt.show()

plt.figure(figsize=(8, 4))

for i in range(quant_segmentos):
    novo_indice = range(quant_geracoes)  
    plt.plot(novo_indice, segmentos[i]['peso_melhor_individuo'], color=colors[i % len(colors)], label=f'Populao {i+1}')

plt.title('Peso do melhor indivduo por Populao')
plt.xlabel('Gerao')
plt.ylabel('Peso')
# plt.xticks(range(quant_geracoes), range(1, quant_geracoes + 1))
plt.show()

plt.figure(figsize=(8, 4))

for i in range(quant_segmentos):
    novo_indice = range(quant_geracoes)  # Criando um novo ndice de 0 a quant_geracoes - 1
    plt.plot(novo_indice, segmentos[i]['valor_melhor_individuo'], color=colors[i % len(colors)], label=f'Populao {i+1}')

plt.title('Valor do melhor individuo por Populao')
plt.xlabel('Gerao')
plt.ylabel('Valor')
# plt.xticks(range(quant_geracoes) / 10, range(10, quant_geracoes ))
plt.show()
```
