```python
import pandas as pd
import matplotlib.pyplot as plt

quant_geracoes = 10
quant_segmentos = 5
df = pd.read_csv('medias.csv')
# segmentos[i] = segmentos[i].head(quant_geracoes)
segmentos = [df.iloc[i:i+quant_geracoes] for i in range(0, len(df), quant_geracoes)]
# print(segmentos[i].describe())

# Criando grficos para cada uma das colunas especificadas

for i in range(quant_segmentos):
    plt.figure(figsize=(8, 4))

    # Grfico para 'media_peso'
    plt.subplot(2, 2, 1)
    plt.plot(segmentos[i]['media_peso'])
    plt.title('Mdia do Peso')
    plt.xlabel('Gerao')
    plt.ylabel('Peso')
    # plt.ylim(0, segmentos[i]['media_peso'].max())

    # Grfico para 'media_valor'
    plt.subplot(2, 2, 2)
    plt.plot(segmentos[i]['media_valor'])
    plt.title('Mdia do Valor')
    plt.xlabel('Gerao')
    plt.ylabel('Valor')
    # plt.ylim(0, segmentos[i]['media_valor'].max())

    # Grfico para 'peso_melhor_individuo'
    plt.subplot(2, 2, 3)
    plt.plot(segmentos[i]['peso_melhor_individuo'])
    plt.title('Peso do Melhor Indivduo')
    plt.xlabel('Gerao')
    plt.ylabel('Peso')
    # plt.ylim(0, segmentos[i]['peso_melhor_individuo'].max())

    # Grfico para 'valor_melhor_individuo'
    plt.subplot(2, 2, 4)
    plt.plot(segmentos[i]['valor_melhor_individuo'])
    plt.title('Valor do Melhor Indivduo')
    plt.xlabel('Gerao')
    plt.ylabel('Valor')
    # plt.ylim(0, segmentos[i]['valor_melhor_individuo'].max())

    plt.subplots_adjust(hspace=0.6, wspace=0.4)
    # Mostrando os grficos
    plt.show()

    # Ajustando layout


```
